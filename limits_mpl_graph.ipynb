```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
mpl.rcParams['font.size']=14
x = np.linspace(0, 1e10, 10000)
y_num = x + 1
y_den = x + 1e6
y = y_num / y_den

plt.plot(x, y_num, 'r-')
plt.plot(x, y_den, 'b--' )
plt.plot(x, y, 'g--')

plt.legend([r'$x +1$', r'$x + 10^6$', r'$fract{x +1}{x + 10^6}$'])

plt.xlabel('x')
plt.ylabel('y')
```

```python
mpl.rcParams['font.size']=14

x = np.linspace(1, 1e-5, 1000)
y_num = 1e-10
y_den = x**2
y = y_num/y_den

plt.plot(y_num, 'ro')
plt.plot(x, y_den, 'b--')
plt.plot(x,y,'g--')
plt.legend([r'$1e-10$', r'$x^2$', r'$\frac{10^{-10}}{x^2}$'])
plt.xlabel('x')
plt.ylabel('y')

```

```python
mpl.rcParams['font.size']=14

x = np.linspace(0, 100, 1000)

y_num = 2*np.exp(x)
y_den = 10*x**4 + 2*x + 2
y = y_num/y_den

plt.plot(x, y_num, 'r-')
plt.plot(x, y_den, 'b--')
plt.plot(x,y, 'g--')
#plt.yscale('log')

plt.legend([r'$2e^{x}$,r'$10x^{4}+2x+2$',r'$\frac{2e^{x}}{10x^{4}+2x+2}$'])
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
```
