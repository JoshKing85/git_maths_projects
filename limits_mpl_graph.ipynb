```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
mpl.rcParams['font.size']=14
x = np.linspace(0, 1e10, 10000)
y_num = x + 1
y_den = x + 1e6
y = y_num / y_den

plt.plot(x, y_num, 'r-')
plt.plot(x, y_den, 'b--' )
plt.plot(x, y, 'g--')

plt.legend([r'$x +1$', r'$x + 10^6$', r'$fract{x +1}{x + 10^6}$'])

plt.xlabel('x')
plt.ylabel('y')
```

```python
mpl.rcParams['font.size']=14

x = np.linspace(1, 1e-5, 1000)
y_num = 1e-10
y_den = x**2
y = y_num/y_den

plt.plot(y_num, 'ro')
plt.plot(x, y_den, 'b--')
plt.plot(x,y,'g--')
plt.legend([r'$1e-10$', r'$x^2$', r'$\frac{10^{-10}}{x^2}$'])
plt.xlabel('x')
plt.ylabel('y')

```

```python
mpl.rcParams['font.size']=12
x=np.linspace(-10, 5, 1000)

y_num = 7*np.power(x, 2) + 14*x + 8 
y_den=x+1
y=y_num/y_den

# Set up the numbers of the x axis
plt.xlim(min(x), max(x))

# Plot the coloured lines
plt.plot(x,y_num,'b-')
plt.plot(x,y_den,'r--')
plt.plot(x,y,'g--')

# Plot the zero line
plt.axhline(y = 0, color = 'k', linestyle = ':')

# Display the legend
plt.legend([r'$7x^2+14x+7$',\
            r'$x+1$',\
    r'$\frac{7x^2+14x+7}{x+1}$'])

# Display the axis labels
plt.xlabel('x')
plt.ylabel('y')
```
