# Limits

```python
import math
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

#### Preliminaries

```python
print(np.finfo(np.float64))
mpl.rcParams['font.size']=12 # Generate x values from 0 to 
# 100000000000 (10 zeros) using 1000 steps
x=np.linspace(0,1e10,1000)
y_num=x+1
y_den=x+1e6
y=y_num/y_den
plt.plot(x,y_num,'b-') #Plot lines
plt.plot(x,y_den,'r--')
plt.plot(x,y,'g--')
# Display the legend. Labels between the $ 
#signs are in latex. The "r" prefix indicates
#Python to expect latex language 
#for the legend text
plt.legend([r'$x+1$',r'$x+10^6$',\
r'$\frac{x+1}{x+10^6}$'])
# Display the axis labels
plt.xlabel('x')
plt.ylabel('y ')#For very large values, numerator and 
#denominator are identical!
#The head-start 10^6 in the denominator
#has no consequences in the limit
```

### a)

```python
mpl.rcParams['font.size']=13 
x=np.linspace(1,1e-5,1000)
x=np.linspace(1,1e-8,1000)
y_num=10**(-10)
y_den=x**2
y=y_num/y_den
plt.axhline(y_num,xmin=min(x),xmax=max(x),c='b')
plt.plot(x,y_den,'r--')
plt.plot(x,y,'g--')
# Display the legend. Labels between the $ 
#signs are in latex. The "r" prefix indicates 
#for the legend text
plt.legend([r'$10^{-10}$',r'$x^2$',\
r'$\frac{10^{-10}}{x^2}$'])
# Display the axis labels
plt.xlabel('x')
plt.ylabel('y')#
```

### b)

```python
#numpy.power better than native python pow
# since we work with arrays

mpl.rcParams['font.size']=14

# Generate a set of values for x
# from -1.1 to -1.00000000001
# x=np.linspace(-1.1, -1.0000000001)
x=np.linspace(-1-1e-1,-1-1e-10,1000)

# Set up the equations
y_num=7*np.power(x,2)+14*x+7 
y_den=x+1
y=y_num/y_den

# Set up the numbers of the x axis
plt.xlim(min(x), max(x))

# Plot the coloured lines
plt.plot(x,y_num,'b-')
plt.plot(x,y_den,'r--')
plt.plot(x,y,'g--')

# Plot the zero line
plt.axhline(y = 0, color = 'k', linestyle = ':')

# Display the legend
# The labels between the $ signs are in latex
plt.legend([r'$7x^2+14x+7$',\
            r'$x+1$',\
    r'$\frac{7x^2+14x+7}{x+1}$'])

# Display the axis labels
plt.xlabel('x')
plt.ylabel('y')

#The numerator goes to zero much faster than the denominator 
# and the limit is zero (ZOOM IN more TO SEE, e.g.
# x=np.linspace(-1-1e-3,-1-1e-15,1000))   
```

### b) Zooming in

```python
# Part a) Zoomed in
#numpy.power better than native python pow
# since we work with arrays

mpl.rcParams['font.size']=12

# Generate a set of values for x
# from -1.1 to -1.00000000001
# x=np.linspace(-1.001, -1.000000000000001)
x=np.linspace(-1-1e-5,-1-1e-10,1000)

# Set up the equations
y_num=7*np.power(x,2)+14*x+7 
y_den=x+1
y=y_num/y_den

# Set up the numbers of the x axis
plt.xlim(min(x), max(x))

# Plot the coloured lines
plt.plot(x,y_num,'b-')
plt.plot(x,y_den,'r--')
plt.plot(x,y,'g--')

# Plot the zero line
plt.axhline(y = 0, color = 'k', linestyle = ':')

# Display the legend
plt.legend([r'$7x^2+14x+7$',\
            r'$x+1$',\
    r'$\frac{7x^2+14x+7}{x+1}$'])

# Display the axis labels
plt.xlabel('x')
plt.ylabel('y')

   
```

### c)

```python
mpl.rcParams['font.size']=12


# Switch y to a log scale on and off
# Now investigate as x approaches infinity
x=np.linspace(0,10,1000)
#x=np.linspace(0,100,1000)
#x=np.linspace(50,200,1000)
# x=np.linspace(690,700,1000)
#x=np.linspace(690,709,1000)

y_num=2*np.exp(x)
y_den=10*np.power(x,4)+2*x+2
y=y_num/y_den

plt.plot(x,y_num,'b-')
plt.plot(x,y_den,'r--')
plt.plot(x,y,'g--')

plt.legend([r'$y=2*e^{x}$',\
            r'$10x^4+2x+2$',\
            r'$y=\frac{2*e^{x}}{10x^4+2x+2}$'])

plt.xlabel('x')
plt.ylabel('y (log-scale)')
plt.yscale('log') 

#The exponential grows much faster
# than the polynomial
```

### d)

```python
mpl.rcParams['font.size']=12

x=np.linspace(0,50,1000)
# x=np.linspace(0,100,1000)
# x=np.linspace(0,120,1000)
# x=np.linspace(0,125,1000)
# x=np.linspace(120,125,1000)
# x=np.linspace(129,130,1000)
# x=np.linspace(129,150,1000)
#x=np.linspace(140,141,1000)

#numpy.power better than native python pow
# since we work with arrays

y_num= (10*np.exp(5*x))-(2*np.exp(-2*x))+np.exp(-x)
y_den=5*np.exp(5*x)+np.exp(-2*x)
y=y_num/y_den

plt.xlim(min(x), max(x))

plt.plot(x,y_num,'b-')
plt.plot(x,y_den,'r--')
plt.plot(x,y,'g--')

plt.axhline(y = 0, color = 'k', linestyle = ':')

plt.legend([r'$10e^{5x}-2e^{-2x}+e^{-x}$',\
            r'$5e^{5x}+e^{-3x}$',\
    r'$\frac{10e^{5x}-2e^{-2x}+e^{-x}}{5e^{5x}+e^{-3x}}$'])

plt.xlabel('x')
plt.ylabel('y (log-scaled)')
plt.yscale('log')
```

### e) ( to the right of 1.0)

```python
mpl.rcParams['font.size']=14

x=np.linspace(1.00001, 1.1, 1000)
# x=np.linspace(1+1e-5,1.1,1000)

y_num=x
y_den=2*np.power(x,2)+2*x-4 
y=y_num/y_den

plt.xlim(min(x), max(x))

plt.plot(x,y_num,'b-')
plt.plot(x,y_den,'r--')
plt.plot(x,y,'g--')

plt.axhline(y = 0, color = 'k', linestyle = ':')

plt.legend([r'$x$', r'$2x^2+2x-4$',\
    r'$\frac{x}{2x^2+2x-4}$'])

plt.xlabel('x')
plt.ylabel('y (log-scaled)')
plt.yscale('log')
#Thus the denominator goes to zero faster than the numerator
# and the limit is +infinity
```

### e) (to the left of 1.0)

```python
# To the left of 1.0

mpl.rcParams['font.size']=14

# x from 0.999 to 0.99999 
# x=np.linspace(1-1e-3,1-1e-5,1000)
x=np.linspace(0.999, 0.99999,1000)


y_num=x
y_den=2*np.power(x,2)+2*x-4 
y=y_num/y_den

plt.xlim(min(x), max(x))

plt.plot(x,y_num,'b-')
plt.plot(x,y_den,'r--')
plt.plot(x,y,'g--')

plt.axhline(y = 0, color = 'k', linestyle = ':')

plt.legend([r'$x$',\
            r'$2x^2+2x-4$',\
    r'$\frac{x}{2x^2+2x-4}$'])

plt.xlabel('x')
plt.ylabel('y')


#y-scale is linear since the numbers are negative
# As x approaches 1 from the left, y tends to negative infinity
# As x approaches 1 from the right, y tends to positive infinity
```
