```python
import math
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

```

**Element 1,  Permutations:** $$ P(n, r) = \frac{n!}{(n - r)!} $$

```python
# The notation P(a, 1) represents the pemuatations of the set, a is equal to number of elements and 1 the group size to choose from.
# A permutation is the set in a particular order.
# The formula for nPr is n!/(n-r)!. n represents the number of elements, r the group size and ! is the operator factorial. The total elements a, is equal to 13. 
# The group size is 1, meaning only one element can be chose from total a. 
# Therefore there are 13 permutations.

# We can check this answer the using formula.
a = 13

perm = int((math.factorial(a))/math.factorial(a - 1))
print(perm)
```

**Element 2,  Combinations:** $$ \binom{n}{r} = \frac{n!}{r!(n - r)!} $$

```python
# The notation (a, 1) represents the combinations of the set, a is equal to number of elements and 1 the group size to choose from
# Combinations are similar to permutations but an important distiction is the selection of an element means it cannot be selected again in a different order. (A, B, C) <=> (A, C, B).
# The formula for nCr is n!/(n-r)!r!. n represents the number of elements, r the group size and is the operator factorial! The total elements a, is equal to 13.
# The group size is 1, meaning only one element can be chosen once from total a. 
# Therefore there are 13 combinations.

# We can check this answer using the math library combinations function using variable a from last cell.

comb = math.comb(a, 1)
print(comb)
```

**Element 3,  Limit:**

$$
\lim_{x \to 0} \frac{a!}{x}
$$


```python
x = np.linspace(-10, 10, 1000)
y_num = math.factorial(a)
y_den = x
y = -(y_num/y_den)

mpl.rcParams['font.size']=14
plt.plot(x, y, 'r--')
plt.xlim(-10, 10)
plt.grid()
plt.axhline(c='black', linewidth=1)
plt.axvline(c='blue', linestyle='--', linewidth=1)
plt.yscale('symlog') 
plt.show()
```
