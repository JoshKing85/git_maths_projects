```python
# always import libraries before use.
import numpy as np
import matplotlib.pyplot as plt
```

# scatter plot

```python
# create or import data to use

x_data = np.random.random(50) * 100
y_data = np.random.random(50) * 100

# pass data into scatter plot cmd, x axis first arg, y axis sec arg. pass style changes in brackets i.e colour(can pass hexdec colour), size and marker 
plt.scatter(x_data, y_data, c= 'red', s= 150, marker = '*' )
# if not in jupyter note book you must type cmd show -> plt.show()
```

# line graph

```python
years = np.array([2005 + x for x in range(5) ])
weight = np.array([83, 84, 85, 83, 87,])
print(years, weight)

# make line thicker and dashed. solid line without linestyle
plt.plot(years, weight, lw=3, linestyle='--'   )
```

# bar plots

```python
#similar to previous graphs but x axis has no value just title
x = ['C++', 'C#', 'Python', 'Java', 'Go']
y = [5, 7, 12, 3, 9]

# style bar width and edge
plt.bar(x, y, color='r', align='edge', width= .75, edgecolor='#000')
```

# histograms

```python
# random data set of ages between 1 and 10. groups are,  5 - (5<= x < 6),  20 -(6<= x <8),  15 -(8 <= x < 10),  10 - ( 10 <= x < 14).
import random
ages = [5, 5, 5, 5, 5]
for x in range(20):
    x = random.randint(6, 7)
    ages.append(x)

for y in range(15):
    y = random.randint(8, 9)
    ages.append(y)

for z in range(10):
    z = random.randint(10,13)
    ages.append(z)

bins_data = [5, 6, 8, 10, 14]
# bins are the indiviual bars, width plotted by the difference of elements in array bins_data.

plt.hist(ages, bins= bins_data, edgecolor = 'black')



# tbc when i know hot to plot frequency density(probaliity).
```

```python
# u can  make  cumalative graph for same data by passing the cumlative cmd and adding a boolean value 
import random
ages = [5, 5, 5, 5, 5]
for x in range(20):
    x = random.randint(6, 7)
    ages.append(x)

for y in range(15):
    y = random.randint(8, 9)
    ages.append(y)

for z in range(10):
    z = random.randint(10,13)
    ages.append(z)

bins_data = [5, 6, 8, 10, 14]

plt.hist(ages, bins= bins_data, cumulative= True)

# all graphs can be headed and axis labled.
# the font can be styled by  cmd fontname and passing a font type lists can be found.
#https://jonathansoma.com/lede/data-studio/matplotlib/list-all-fonts-available-in-matplotlib-plus-samples/
plt.title('Made up data for tutorial', fontname='Gabriola' )
plt.xlabel('Ages', fontname= 'Arial')
plt.ylabel('number of occurances', fontname= 'MS Gothic')

```

# pie charts

```python
langs = ['Python', 'C#', 'C++', 'Java', 'Go']

votes = np.array([40, 24, 30, 14, 20 ])
# percetages show up to decimal place using float data type. without pctdistace % will be inside of chart

plt.pie(votes, labels=langs, autopct = '%.2F%%', pctdistance= 1.3,)
```
