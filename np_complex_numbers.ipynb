# Plot Argand diagram

```python
z = 1 + 2j
plt.plot(z.real, z.imag, 'ro', markersize=5,)
# add text at coordinates, but increase y axis plot so text is higer.
plt.text(z.real,(z.imag+0.1*z.imag), z )

plt.xlim(0,3);plt.ylim(0,3)
# shows  point plotted with vector from origin
plt.quiver(0, 0, z.real, z.imag, angles='xy', scale_units='xy', scale=1, color='blue')
# label axis
x_label = 'Re'
y_label = 'Img'
plt.xlabel(x_label), plt.ylabel(y_label)
plt.title('Argand Diagram')
```
