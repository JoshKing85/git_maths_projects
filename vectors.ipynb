```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
w = np.array([1,1])
alphas2 = [3, 2, 1, 0, 5, 0]
for i, value in enumerate(alphas2):
    alpha = np.sqrt(value)
    v = w*alpha
    plt.plot([0, v[0]], [0, v[1]], label=f'alpha {i + 1}: {alpha:.2f}')
    plt.arrow(0, 0, v[0], v[1], head_width=0.05, head_length=0.05)
    print('vector', v)
    print('length for alpha', alpha, 'is', np.linalg.norm(v))

    plt.legend(fontsize=10, loc= 'upper left')
```
